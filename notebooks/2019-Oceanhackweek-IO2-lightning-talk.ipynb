### Interactive Oceans 2.0: Data Portal and Data Access Tools

Who we are: UW Cabled Array Value Add Project

What we do: The goal of CAVA is to increase the active use of Regional Cabled Array data by scientists, with vision to support educators and public exploration of data in the future.

What we have created so far:
- Python library for OOI Data Access: [`yodapy`](https://github.com/cormorack/yodapy)
- Cloud copy of Shallow Profiler and Platform data in [`io2data`](https://io2data.s3.us-west-2.amazonaws.com/index.html) S3 Bucket in Zarr Format
- [Interactive Oceans Data Portal](https://app.interactiveoceans.washington.edu/)

## `yodapy`

```python
>>> from yodapy.datasources import OOI
>>> ooi = OOI()
>>> ooi.search(region='cabled', site='axial base shallow profiler', node='shallow profiler', instrument='CTD')
>>> ooi.view_instruments()
>>> begin_date = '2018-01-01'
>>> end_date = '2018-01-02'
>>> ooi.request_data(begin_date=begin_date, end_date=end_date)
>>> ds_list = ooi.to_xarray()
```
